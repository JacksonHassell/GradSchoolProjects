```python
!pip install selenium
!apt-get -q update   #Used to handle installation and removal of softwares and libraries
!apt install -yq chromium-chromedriver #ChromeDriver is a separate executable that Selenium WebDriver uses to control Chrome.
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
#WebDriver is a browser automation framework that works with open source APIs.
#The framework operates by accepting commands, sending those commands to a browser, and interacting with applications.
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
wd = webdriver.Chrome('chromedriver',options=chrome_options)
```

```python
baseUrl = "https://www.yelp.com/search?find_desc=Restaurants&find_loc=Austin%2C+TX"
wd.get(baseUrl)

elements = wd.find_elements_by_class_name('css-og60gk')[2:]
links = [e.get_attribute('href') for e in elements]
names = [e.text for e in elements]

for i in range(10, 30, 10):
  url = baseUrl + "&start=" + str(i)
  wd.get(url)
  elements = wd.find_elements_by_class_name('css-og60gk')[2:]
  current_links = [e.get_attribute('href') for e in elements]
  current_names = [e.text for e in elements]
  links.append(current_links)
  names.append(current_names)
  print(i)
print(len(links))
```

```python
print(links)
```

```python
product_name = []
product_review = []
user_rating = []

for link in links:
  wd.get(link)
  scores = wd.find_elements_by_class_name('BAscore_norm')
  name = wd.find_elements_by_xpath('/html/body/div[1]/div/div[2]/div[2]/div[2]/div/div/div[3]/div/div/div[1]/h1')[0].text
  for score in scores:
    user_rating.append(score.text)
    product_name.append(name)
  reviews = wd.find_elements_by_class_name('rating_fullview_content_2')
  for review in reviews:
    product_review.append(review.text)
  break

print(product_name)
print(product_review)
print(user_rating)
```
